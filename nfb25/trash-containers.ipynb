```python
import pandas as pd

df = pd.read_csv('trash-containers.csv')
df = df[['neighborhood']]
df = df['neighborhood'].value_counts()
data = {
  "neighborhood": df.keys(),
  "count": df.values
}

df = pd.DataFrame(data)
df.to_csv('trash-containers-formatted.csv', index=False)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
plt.xlabel('Neighborhood')
plt.ylabel('Containers')
plt.bar(df['neighborhood'], df['count'], width = .6, color='skyblue')
plt.title('Smart Trash Containers Per Neighborhood')
plt.xticks(rotation=45, ha='right')
plt.xticks(fontsize=8)
plt.tight_layout()
plt.show()
```
